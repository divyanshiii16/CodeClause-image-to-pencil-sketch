**Converting RGB images to Grayscale image using OpenCV**

```python
# importing OpenCV library
import cv2
```

```python
img=cv2.imread('/content/dog.854dd8955515c408033b24db7f2a6d1a?rik=EbmhZJEVVX6hTQ&riu=http:%2F%2Fwww.cutenessoverflow.com%2Fwp-content%2Fuploads%2F2014%2F02%2FLittle-Cute-Puppy-1600x1200.jpg&ehk=MPYGeeHh59OIiNzQOYgdPoFwCvBH1Ubp%2FXKtPv2NK%2FU=&risl=&pid=ImgRaw&r=0')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

**Converting into sketch**


```python
invert = cv2.bitwise_not(grayscale_image)
```

```python
blur = cv2.GaussianBlur(invert,(21,21),0)
```

```python
invertedblur = cv2.bitwise_not(blur)
```

```python
sketch_filter = cv2.divide(grayscale_image,invertedblur,scale=249.0)
```

```python
#displaying the image

cv2_imshow(sketch_filter)
```

```python
cv2.imwrite('sketch.jpg',sketch_filter)
```
